```python
class HotelManagment:
   
    def __init__(self,room_no='',name='',num_of_rooms='',duration=''):
        self.name=name
        self.num_of_rooms=num_of_rooms  # number of rooms user wants to book
        self.room_no=room_no
        self.duration=duration #duration of his stay
    def booking(self):
        self.name=input('enter ur name:')
        self.num_of_rooms=int(input('enter number of rooms you want:'))
        self.room_no=input('enter room number assigned:')
        self.duration=int(input('enter duration of your stay:'))
    
    def display(self):
        print('')
        print('room no:', self.room_no)
        print('name of person living in room:', self.name)
        print('Total number of room other than this are booked by him:', self.num_of_rooms-1)
        print('Duration he will stay:', self.duration)
```

```python
def delet(rng,list1):
    i=0  # for loop iteration
    print('enter from options')
    print('1.name of customer u want to delete')
    print('2.room number u want to delete')
    print('3.Exit from delete option')
    print('enter the option')
    choice1=input() 
    if(choice1=='1'):
        dummy=0
        print('enter the name of customer')
        nam=input()
        while i<rng:
         if list1[i].name==nam:
            list1[i].display()  #each index in the list has a adress checking all adresses
            list1.pop(i)
            print('deleted!!!')
            dummy=1
            break
         i=i+1
        if(dummy==0):
           print('no details available for the entered value')
    elif(choice1=='2'):
        dummy=0
        print('enter the room number')
        no=input() # room number input
        while i<rng:
         if list1[i].room_no==no:
            list1[i].display()
            list1.pop(i)
            print('deleted!!!')
            dummy=1
            break
         i=i+1
        if(dummy==0):
            print('no details available for the entered value')
    elif(choice1=='3'):
        print('exiting from delete menu....................')
    else:
        print('enter a valid choice:')
```

```python
def edit(rng,list1):
    i=0  # for loop iteration
    print('enter from options')
    print('1.name of customer for whom u want to edit the details?')
    print('2.which room number details u want to edit?')
    print('3.Exit from edit option')
    print('enter the option')
    choice1=input()
    if(choice1=='1'):
        dummy=0
        print('enter the name of customer')
        nam=input()   #name of customer
        while i<rng:  #each index in the list has a adress of object checking all adresses
         if list1[i].name==nam:  
            print('---Enter all details again----')
            list1[i].booking()
            print('done editing')
            dummy=1
            break
         i=i+1
        if(dummy==0):
           print('no details available for the entered value')
    elif(choice1=='2'):
        dummy=0
        print('enter the room number')
        no=input()   #room number
        while i<rng:
         if list1[i].room_no==no:
            print('---Enter all details again----')
            list1[i].booking()
            print('done editing')
            dummy=1
            break
         i=i+1
        if(dummy==0):
           print('not available')
    elif(choice1=='3'):
        print('exiting from edit menu....................')
    else:
        print('no details available for the entered value')
```

```python
def enquiry(rng,list1,total_rooms):
    i=0  # for loop iteration
    print('enter from options')
    print('1.Searching with name of customer')
    print('2.Searching with room number')
    print('3.Check for room availability ')
    print('4.Exit from enquiry option')
    print('enter ur choice')
    choice1=input()
    if(choice1=='1'):
        dummy=0
        print('enter the name of customer')
        nam=input()
        while i<rng: #each index in the list has a adress of object checking all adresses
         if list1[i].name==nam:  # if name avialable in list then enter this
            print('available')
            list1[i].display()
            dummy=1
            break
         i=i+1
        if(dummy==0):
           print('not available')
    elif(choice1=='2'):
        dummy=0
        print('enter the room number')
        no=input()
        while i<rng:
         if list1[i].room_no==no:
            print('available')
            list1[i].display()
            dummy=1
            break
         i=i+1
        if(dummy==0):
           print('not available')
    elif(choice1=='3'):
        sum1=0
        while i<rng: # iterated through whole list summation of number of rooms by each customer was done
          sum1=sum1+list1[i].num_of_rooms  
          i=i+1
        if sum1==total_rooms:  # if total rooms equal to this sum then no room available
            print('No more rooms available')
        else:
            available=total_rooms-sum1
            print('total rooms available:',available)
    elif(choice1=='4'):
        print('exiting from enquiry menu....................')
    else:
        print('enter a valid choice:')
```
